### Assignment 
Create class triangle with 3 sides

1. Does from a triangle : Sum of any two sides be greater than third
2. Perimeter : Sum of all sides of a triangle
3. Area : Herons formula

#### Create class triangle

```python
from dataclasses import dataclass
from typing import Union
```

```python
@dataclass
class Triangle:
    a: Union[int,float]
    b: Union[int,float]
    c: Union[int,float]

    def __post_int__(self):
        errors = []

        if not isinstance(self.a,(int,float)) or self.a <= 0:
            errors.append("Side a should be a positive integer or float ")
        if not isinstance(self.b, (int,float)) or self.b <= 0:
            errors.append("Side b should be a positive integer or float")
        if not isinstance(self.c, (int,float)) or self.c <= 0:
            errors.append("Side c should be a positive integer or float")
        if not self.is_triangle():
            errors.append("Given sides do not form a triangle")

        if errors:
            raise ValueError("\n".join(errors))
        
    def is_triangle(self):
        if (
            (self.a + self.b > self.c)
            and (self.b + self.c > self.a)
            and (self.a + self.c > self.b)

        ):
            return True
        else:
            return False
        
    def perimeter(self):
        return self.a + self.b + self.c
    
    def area(self):
        s= self.perimeter()/2
        a= (s*(s-self.a)*(s-self.b)*(s-self.c))**(1/2)
        return a
```

### Polymorphism in class

```python
class India:
    def capital(self):
        print("New Delhi is capital of India")

    def language(self):
        print("Hindi is widely spoken in India")
```

```python
class USA:
    def capital(self):
        print("Washington D.C. is capital of USA")

    def language(self):
        print("English is widely spoken in USA")
        
```

```python
class France:
    def capital(self):
        print("Paris is the capital of France")
    def language(self):
        print("French is widely spoken in France")

```

#### Practical application - Calculation area or perimeter of multiple shapes

```python
@dataclass
class Reactangle:
    width: Union[int,float]
    height: Union [int,float]

    def perimeter(self):
        return 2 * (self.width + self.height)
    def area(self):
        return self.width * self.height
```

```python
@dataclass
class Circle:
    radius: Union[int,float]

    def perimeter(self):
        return 2 * pi * self.radius
    def area(self):
        return pi*(self.radius**2)
```
